```python
sql = """
DROP TABLE IF EXISTS raw_orders;

CREATE TABLE raw_orders (
    order_id INT,
    customer_email VARCHAR(100),
    order_date DATE,
    amount DECIMAL(10,2),
    status VARCHAR(20)
);

INSERT INTO raw_orders VALUES
(1, 'alice@email.com', '2024-01-10', 500.00, 'completed'),
(2, 'alice@email.com', '2024-01-10', 500.00, 'completed'),  -- Duplicate!
(3, 'bob@email.com', '2024-01-12', 15000.00, 'completed'),   -- Suspiciously high
(4, 'carol@email.com', '2024-01-14', -100.00, 'completed'),  -- Negative amount!
(5, 'dave@email.com', '2024-01-15', 200.00, 'completed');


"""
```

#### with insert/create command we commit 

```python

with get_connection() as connection:
    with connection.cursor() as cursor:
        cursor.execute(sql)
        connection.commit()   # REQUIRED for CREATE / INSERT
```
